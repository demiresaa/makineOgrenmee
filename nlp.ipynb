```python
def islem (review):
    review=BeautifulSoup(review).get_text()
    review=re.sub("[^a-zA-Z]"," ",review)
    review.lower()
    review=review.split()
    swords=set(stopwords.words("english"))
    review=[w for w in review if w not in swords]
    return(" ".join(review))
    # burada ise ornek_metin için yaptığımız işlerim butun dataya yaptık
```

```python
train_x_tum=[]
for i in range(len(data["review"])):
    if(i+1)%1000==0:
        print("işlem adımı",i+1)
    train_x_tum.append(islem(data["review"][i]))
    # burada yapılan işlemi tek tek her bir metne yaptır ve yeni dizeye aktar dedik
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=train_x_tum
y=np.array(data["sentiment"])
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=33)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vektorizer=CountVectorizer(max_features=5000)  # en fazla 5000 kelime özellik olsun dedik ve bunlaraı sayısallaştırdık
```

```python
train_x1=vektorizer.fit_transform(train_x) # dönüştürdük sayısala
train_x1
```

```python
train_x1=train_x1.toarray() # daha iyi ve hatasız çalışması için arraya dönüştürdük
```

```python
train_y1=train_y
```

```python
train_x1.shape,train_y1.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100)
```

```python
rf.fit(train_x1,train_y1)
```
