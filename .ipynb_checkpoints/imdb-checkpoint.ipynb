```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("C:/Users/Dell/Desktop/makineOgrenme/users.data",sep="\t",names=["user_id","item_id","rating","timestamp"])
```

```python
data.head()
```

```python
movie=pd.read_csv("C:/Users/Dell/Desktop/makineOgrenme/movie_id_titles.csv")
```

```python
movie.head()
```

```python
data=pd.merge(data,movie,on="item_id") 
# burada datanın item_id si ile movienin item_idleri üzerinden aynı olanları birleştirdik
```

```python
data.head()
```

```python
tablo=data.pivot_table(index="user_id",columns="title",values="rating")
tablo.head()
```

```python
starwars=tablo["Star Wars (1977)"]
starwars.head()
```

```python
korelasyon=tablo.corrwith(starwars) # starwarsın korelasyonlarına baktık
```

```python
korelasyon
```

```python
kore=pd.DataFrame(korelasyon,columns=["Correlation"])
kore.dropna(inplace=True)
kore.head() # korelasyondaki nan değerlerini sildik
```

```python
kore.sort_values("Correlation",ascending=False).head(15) # burada korelasyonları bbüyükten küçüğe doğru sıraladık ama oy sayisi bizi şaşırtacağı için bu sırlama önemli değil
```

```python
data.drop((["timestamp"]),axis=1)
```

```python
ratings=pd.DataFrame(data.groupby("title")["rating"].mean()) # burada her kullanının film için verdiği ratinglerin ortsi var
ratings.sort_values("rating",ascending=False).head()
```

```python
ratings["oysayisi"]=pd.DataFrame(data.groupby("title")["rating"].count())
```

```python
ratings.head()
```

```python
ratings.sort_values("oysayisi",ascending=False).head()
```

```python
kore.head()
```

```python
kore=kore.join(ratings["oysayisi"])
kore.head()
```

```python
kore[kore["oysayisi"]>100].sort_values("Correlation",ascending=False).head()
```
